```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'C', 'JavaScript', 'HTML/CSS', 'SQL', 'PHP', 'Jupyter Notebook']
sizes = [25, 18, 12, 10, 8, 8, 8, 5, 6]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#c2f0c2','#ffb366','#c299ff']

plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Programming Languages & Tools Usage', pad=20)
plt.axis('equal')
plt.savefig('languages_pie_chart.png')  # Saves in current directory
plt.show()
```
